# 05 Final Load Prep

Use this notebook to compute final KPIs, prepare the Tableau-ready dataset, and export the exact file used in the dashboard.

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().resolve().name == 'notebooks' else Path.cwd().resolve()
```

```python
DATA_PATH = PROJECT_ROOT / 'data/processed/cleaned_credit_data.csv'
TABLEAU_READY_PATH = PROJECT_ROOT / 'data/processed/tableau_ready_dataset.csv'

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
# Add KPI calculations or aggregations here.
df["total_bill_amt"] = df[["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]].sum(axis=1)
df["total_pay_amt"] = df[["PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]].sum(axis=1)
df["pay_to_bill_ratio"] = df["total_pay_amt"] / df["total_bill_amt"].replace(0, 1)
df["age_group"] = pd.cut(df["AGE"], bins=[20, 30, 40, 50, 60, 100], labels=["20-30", "30-40", "40-50", "50-60", "60+"])
df.head()
```

```python
TABLEAU_READY_PATH.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(TABLEAU_READY_PATH, index=False)
print(f'Saved Tableau-ready dataset to {TABLEAU_READY_PATH}')
```
