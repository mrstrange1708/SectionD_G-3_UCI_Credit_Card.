## 7. Check for Duplicate Rows
Checking for exact duplicate rows in the dataset. Expected: 0.

```python
print(f'Duplicate rows: {df.duplicated().sum()}')
```

## 8. Value Counts on Categorical Columns
Running value counts on specific categorical columns where dirty/undocumented values might live.
- **EDUCATION**
- **MARRIAGE**
- **SEX**
- **default.payment.next.month**

```python
cols_to_check = ['EDUCATION', 'MARRIAGE', 'SEX', 'default.payment.next.month']
for col in cols_to_check:
    print(f'\n--- Value counts for {col} ---')
    print(df[col].value_counts(dropna=False))
```

## 9. Class Imbalance Check
Calculating what percentage of rows are default=1 vs default=0.

```python
default_counts = df['default.payment.next.month'].value_counts(normalize=True) * 100
print('Class Distribution (Percentage):') 
print(default_counts)
```

> **Observation**: The dataset exhibits a class imbalance. Approximately 77.88% of the clients did not default (class `0`), while 22.12% defaulted (class `1`).
